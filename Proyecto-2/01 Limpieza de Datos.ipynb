```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None) # MUESTRA TODAS LAS COLUMNAS DE UN PANDAS DATAFRAME
from fastapi import FastAPI
import ast
import sklearn
```

```python
# Cargo y uno los datasets
d1= pd.read_csv(r'C:\Users\Edwin\Desktop\Henry\M7. Proyecto Individual I\Datasets\movies_dataset.csv', sep=',')
d2= pd.read_csv(r'C:\Users\Edwin\Desktop\Henry\M7. Proyecto Individual I\Datasets\credits.csv',sep=',')
data= pd.concat([d1, d2], axis=1)
data.head()
```

```python
# Eliminar las columnas que no serán utilizadas,
# video,imdb_id,adult,original_title,poster_path y homepage.
columnas_no_utilizadas = ['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage']
data.drop(columnas_no_utilizadas, axis=1, inplace=True)
```

```python
#Detectar valores NaN con la función isna() y sumarlos por campo
data.isna().sum()
```

```python
# Rellenar los valores nulos en los campos 'revenue' y 'budget' con 0
data['revenue'].fillna(0, inplace=True)
data['budget'].fillna(0, inplace=True)
```

```python
# Eliminar filas con valores nulos en 'release date'
data.dropna(subset=['release_date'], inplace=True)
```

```python
# Ver si quedaron valores nulos que revisar 
data.isna().sum()
```

```python
data.info()
```

```python
# Convertir la columna 'release date' a tipo de datos de fecha
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')

# Cambiar las fechas en el formato 'AAAA-mm-dd'
data['release _date'] = data['release_date'].dt.strftime('%Y-%m-%d')
```

```python
# Crear la columna 'release_year' a partir del año de la fecha de estreno
data['release_year'] = data['release_date'].dt.strftime('%Y')
```

```python
data.head(4)
```

```python
# Verifico si se cambiaron los tipos de datos
data.info()
```

```python
# Convertir las columnas 'revenue' y 'budget' a tipo numérico
data['revenue'] = pd.to_numeric(data['revenue'], errors='coerce')
data['budget'] = pd.to_numeric(data['budget'], errors='coerce')

# Crear la columna 'return' con el cálculo del retorno de inversión y tratar los valores de cero
data['return'] = np.where(
    (data['budget'] == 0) | (data['revenue'] == 0),
    0,  # Valor predeterminado para los casos en los que budget o revenue sea cero
    data['revenue'] / data['budget'])
```

```python
# Verificar si los valores nulos fueron reemplazados por el valor numérico 0
data.info()
```

```python
# Eliminar la primera columna "release date"
data.drop('release_date', axis=1, inplace=True)
```

```python
# Convertir la columna 'belongs_to_collection' en formato de diccionario
# Es normal que este proceso tarde 6 min o mas
data['belongs_to_collection'] = data['belongs_to_collection'].apply(lambda x: eval(x) if pd.notnull(x) else {})
data['genres'] = data['genres'].apply(lambda x: eval(x) if pd.notnull(x) else {})
data['spoken_languages'] = data['spoken_languages'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
data['production_companies'] = data['production_companies'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) and isinstance(x, str) else [])
data['production_countries'] = data['production_countries'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
data['cast'] = data['cast'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
data['crew'] = data['crew'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])

# Desanidar 
df_production_countries = data['production_countries'].apply(lambda x: [country['name'] for country in x] if isinstance(x, list) else [])
df_belongs_to_collection = data['belongs_to_collection'].apply(lambda x: x.get('name', '').replace(' Collection', '') if isinstance(x, dict) else '')
df_genres = data['genres'].apply(lambda x: [genre['name'] for genre in x])
df_spoken_languages = data['spoken_languages'].apply(lambda x: [lang['name'] for lang in x])
df_production_companies = data['production_companies'].apply(lambda x: [prod['name'] for prod in x] if isinstance(x, list) else [])
df_cast = data['cast'].apply(lambda x: [(item['character'], item['name']) for item in x])
df_crew = data['crew'].apply(lambda x: [(item['department'], item['name']) for item in x])
df_directors = data['crew'].apply(lambda x: [item['name'] for item in x if item['department'] == 'Directing'])

duplicated_columns = ['belongs_to_collection', 'genres', 'spoken_languages', 'production_companies', 'production_countries', 'cast', 'crew']

# Crear el DataFrame df_movies con las columnas desanidadas y las demás columnas del dataset original
df_movies = pd.concat([df_belongs_to_collection, df_genres, df_spoken_languages, df_production_companies, df_production_countries, df_cast, df_crew.rename('crew'), df_directors.rename('directors'), data.drop(columns=duplicated_columns)], axis=1)
```
